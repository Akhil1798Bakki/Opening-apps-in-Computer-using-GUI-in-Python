### The main aim of this project is to open few files by using python platform
### In this project we will be able to open MS Word, MS Power Point, MS Excel, Google Chrome, Whats App by using python GUI
### The location of the specific .exe files are according to my pc location and it can be varied depending on Desktops/Mac
### In this project we have defined various classes for various functions of MS Word, MS Power point, MS Excel, Google Chrome, Whats App
### In this project each class is called by the button through the command of class initialized in the button
### This code helps us to understand how can we take help of python in openning various applications in our PC/Mac

```python
from tkinter import *# Initializing tkinter for building GUI
from tkinter import filedialog, Text # this is used for importing various .exe files
import os # Importing OS for executing the .exe files

root=Tk() # Initializing the Tk
root.configure(bg='#424242')# Giving the background colour for GUI
root.geometry('450x500') # Specifying the geometry of the GUI
apps = [] # Initializing the apps

# The below class is used for opeining MS Word
class Word:
    def __init__(self, master):
        self.master = master
    def createView(self):
        for widget in frame.winfo_children():
            widget.destroy()
        
        filename = filedialog.askopenfilename(initialdir = "C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Word", title = "Word doc", filetypes = (("Executables", "*.exe"),("All Files", "*.*"))) # The file location of MS Word can be varied depending on system
        apps.append(filename)
        print(filename)
        for app in apps:
            label = Label(frame, text = app, bg = "gray").pack()
word = Word(root)

# The below class is used for opening MS Power Point
class powerPoint:
    def __init__(self,master):
        self.master = master
    def createView(self):
        for widget in frame.winfo_children():
            widget.destroy()
        filename = filedialog.askopenfilename(initialdir = "C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Power Point", title = "Power Point", filetypes = (("Executables", "*.exe"),("All Files", "*.*")))# The file location of MS Power Point can be varied depending on system
        apps.append(filename)
        print(filename)
        for app in apps:
            lable = Label(frame, text = app, bg = "gray").pack()
powerpoint = powerPoint(root)

# The below class is used for opening MS Excel
class excelFile:
    def __init__(self,master):
        self.master = master
    def createView(self):
        for widget in frame.winfo_children():
            widget.destroy()
        filename = filedialog.askopenfilename(initialdir = "C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Excel File", title = "Excel file", filetypes = (("Executables", "*.exe"),("All Files","*.*")))# The file location of MS Excel can be varied depending on system
        apps.append(filename)
        print(filename)
        for app in apps:
            lable = Label(frame, text = app, bg = "gray").pack()
excelfile = excelFile(root) 

# The below class is used for opening Google Chrome
class GoogleChrome:
    def __init__(self,master):
        self.master = master
    def createView(self):
        for widget in frame.winfo_children():
            widget.destroy()
        filename = filedialog.askopenfilename(initialdir = "C:\\ProgramData\\Microsoft\\Windows\\Start Menu\\Programs\\Google Chrome", title = "Google Chorme", filetypes = (("Executables", "*.exe"),("All Files","*.*")))# The file location of Google Chrome can be varied depending on system
        apps.append(filename)
        print(filename)
        for app in apps:
            lable = Label(frame, text = app, bg = "gray").pack()
googlechrome = GoogleChrome(root)

# The below class is used for opening Whats App
class WhatsApp:
    def __init__(self,master):
        self.master = master
    def createView(self):
        for widget in frame.winfo_children():
            widget.destroy()
        filename = filedialog.askopenfilename(initialdir = "C:\\Users\\Akhil\\AppData\\Roaming\\Microsoft\\Windows\\Start Menu\\Programs\\WhatsApp", title = "Whats App", filetypes = (("Executables","*.exe"),("All Files","*.*")))# The file location of Whats App can be varied depending on system
        apps.append(filename)
        print(filename)
        for app in apps:
            lable = Label(frame, text = app, bg = "gray").pack()
whatsapp = WhatsApp(root)

# The below class is used to run all the apps opened by the user
class runApps:
    def __init__(self, master):
        self.master = master
    def createView(self):
        for app in apps:
            os.startfile(app)
runapps = runApps(root)

frame = Frame(root, bg = "White")# This gives the colour of the frame in GUI in root
frame.place(relwidth = 0.8, relheight = 0.6, relx = 0.1, rely = 0.3)# This gives the placement of frame in the window

wordFile = Button(root, text = "Word file", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = word.createView).grid(row = 1, column = 3)# This is the Button used for MS Word
powerPoint = Button(root, text = "Power Point", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = powerpoint.createView).grid(row = 1, column = 4)# This is the Button used for MS Power Point
ExcelFile = Button(root, text = "Excel File", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = excelfile.createView).grid(row = 1, column = 5)# This is the Button used for MS Excel
GoogleChorme = Button(root, text = "Google Chrome", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = googlechrome.createView).grid(row = 1, column = 6)# This is the Button used for Google Chrome
WhatsApp = Button(root, text = "Whats App", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = whatsapp.createView).grid(row = 1, column = 7)# This is the Button used for Whats App
runApps = Button(root, text = "Run file", padx = 10, pady = 5, fg = "white", bg = "#263D42", command = runapps.createView).grid(row = 2, column = 5)# This is the Button used for running all the opened .exe files

root.mainloop()
```
